## Objective

The objective of this notebook is to perform thorough exploratory data analysis. To accomplish this, we create a dataframe of all of the shoes and their sales and label them with the shoe ID and the shoe name. This allows for masks to filter for select shoes, grab their sales and continue down the pipeline workflow smoothly.

```python
import pandas as pd
import numpy as np
import regex as re
import json
import csv
import datetime
import seaborn as sns
import warnings
import matplotlib.pyplot as plt

%matplotlib inline
```

## Import Jordans Dataframe

```python
df_jordans = pd.read_csv('../data/jordan_ids.csv').drop('Unnamed: 0', axis=1)
```

## Create dictionary of Id and Slug

```python
jordan_dict = df_jordans[['product_id', 'slug']].to_dict()
```

## Create dataframe of all shoe sales with corresponding id and slug

Below, instantiate empty lists for our dataframe features. Add all the necessary features and align the product ID and slug to the sales of each shoe.

```python
date = []
saleprice = []
size = []
shoe_id = []
slug = []

for jordan_index in jordan_dict['product_id']:
    filename = jordan_dict['product_id'][jordan_index]
    files = !ls ../data/raw_json/'{filename}'*

    for file in files:
        with open(file, 'r') as f:
            curr_json = json.load(f) 
        for shoe in curr_json:
            date.append(shoe['createdAt'])
            saleprice.append(shoe['amount'])
            size.append(shoe['shoeSize'])
            shoe_id.append(jordan_dict['product_id'][jordan_index])
            slug.append(jordan_dict['slug'][jordan_index])
```

```python
df_sales = pd.DataFrame(
        {'date':date,
         'saleprice':saleprice,
         'size':size,
         'id':shoe_id,
         'slug':slug,
        }
    )
```

### Use Pointers to Shoe ID and establish as new ID

Compress the data slightly by changing datatypes to category dates to create a new feature for the ID and essentially assign pointers to the long unique product IDs.

```python
df_sales['id'] = df_sales['id'].astype('category')
df_sales['slug'] = df_sales['slug'].astype('category')
```

```python
new_id_dict = dict(zip(df_sales['id'].unique(), range(len(df_sales['id'].unique()))))
```

```python
df_sales['new_id'] = df_sales['id'].map(lambda x: new_id_dict[x])
```

## Preprocessing

Clean up the date feature by removing the timezone, changing to datetime type and setting to index.

```python
df_sales['date'] = [re.findall(r'[0-9]{4}-[0-9]{2}-[0-9]{2}', date)[0] for date in df_sales['date']]
```

```python
df_sales['date'] = pd.to_datetime(df_sales['date'])
```

```python
df_sales['date'] = df_sales['date'].apply( lambda df_sales : datetime.datetime(year=df_sales.year, month=df_sales.month, day=df_sales.day))
```

```python
df_sales.set_index(df_sales['date'], inplace=True)
```

```python
df_sales.drop('date', axis=1, inplace=True)
```

## Exploratory Data Analysis

Conduct exploratory data analysis on a sneaker to demonstrate possible trends to take note of.

### Summary Statistics

The mode of distribution of the shoesizes is most center in between sizes 8 and 13.

### Shoe Size vs Price

```python
def scatterPlot(df):
    plt.figure(figsize=(15,8))
    plt.scatter(df['size'], df['saleprice'], color='#5499C7')
    plt.title(f'Jordan Bred 1 2016', fontsize=25)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.xlabel(f'Shoe Size', fontsize=18)
    plt.ylabel('Saleprice ($)', fontsize=18)
    plt.grid()
    plt.show();
```

```python
scatterPlot(df_sales)
```

The saleprice of shoes is most highest for sizes between 8 and 14.

The plot belows is color labeled so that the most lightest colors closest to white are the most highest in number of shoes. The count of the shoe sales are grouped and sorted by size.

```python
counts = df_sales.groupby('size')[['id']].count().to_dict('list')['id']
```

```python
sizes = sorted(df_sales['size'].unique())
```

```python
greys = [str((val-min(counts))/max(counts)) for val in counts] 
```

```python
labels = sorted(df_sales['size'].unique())

fig = plt.figure(figsize=(20,10))

bplot = plt.boxplot([df_sales[df_sales['size'] == size]['saleprice'].values for size in sizes],
                         vert=True, 
                         patch_artist=True,
                         labels=labels)

colors = greys
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.title(f'Jordan Bred 1 2016', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel(f'Shoe Size', fontsize=18)
plt.ylabel('Saleprice ($)', fontsize=18)

plt.show()
```

There are the most shoes of the higher price between size 8 and 13.

```python
def boxenPlot(df):
    plt.figure(figsize=(20,8))
    plt.title(f'Jordan Bred 1 2016', fontsize=25)
    ax = sns.boxenplot(df['size'], df['saleprice'])
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.xlabel(f'Shoe Size', fontsize=18)
    plt.ylabel('Saleprice ($)', fontsize=18)
    
    norm = plt.Normalize(df_sales['saleprice'].min(), df_sales['saleprice'].max())
    sm = plt.cm.ScalarMappable(norm=norm)
    sm.set_array([])
    ax.figure.colorbar(sm)
    
    plt.show();
```

```python
boxenPlot(df_sales)
```

The shoe sizes of between 8 and 9.5 have the highest sale price.

This plot displays the weekly average sales of the shoe since release date.

```python
def basicPlot(df, method):
    if method == 'W':
        time_span = 'Weekly'
    elif method == 'D':
        time_span = 'Daily'
    plt.figure(figsize=(15, 8))
    plt.plot(df['saleprice'].resample(method).mean().fillna(df['saleprice'].mean()), color='#5499C7')
    plt.title(label = f'Jordan Bred 1 2016 Average {time_span} Sales', fontsize=25)
    plt.xticks(fontsize=15, rotation=90)
    plt.yticks(fontsize=15)
    plt.xlabel(f'Date', fontsize=18)
    plt.ylabel('Saleprice ($)', fontsize=18)
    plt.grid()
    plt.show();
```

```python
basicPlot(df_sales, 'W')
```

The plot below displays the daily average sale since release date.

```python
basicPlot(df_sales, 'D')
```
