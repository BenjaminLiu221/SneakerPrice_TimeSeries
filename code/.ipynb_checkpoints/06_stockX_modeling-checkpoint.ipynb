## Objective

The objective of this notebook is to forecast time series of historic saleprices data of a sneaker. First a baseline persistence forecast model will be used to predict saleprice and then compared to the exploration of long short term memory neural networks such that we enter tensor (input) into a network hidden layer and exports output and the models will be evaluated through metric root mean squared error. Plots will be used to visualize the predictions as well as evaluate if the model makes sense. In addition, a multiclass classification logistic regression model will used to recommend a user depending on the past historic sales data, whether the user should be looking to sell the particular sneaker, hold off on buying the sneaker, or buying the sneaker at the points of forecasted price. The accuracy score metrics will be used to evaluate the model.

## Persistence Model Forecast

The baseline forecast used for this time series is a persistence forecast where the observation from the prior time step $t_{-1}$ is used to predict the observation at the current time step $t_1$. We can take the last observation from the training data and history accumulated by walk-forward validation and use that to predict the current time step.

```python
df.plot()
plt.title('Jordan Bred 1 Weekly Sales')
plt.ylabel('Price ($)')
plt.xlabel('Date')
```

There is an increasing upwards trend in saleprice.

### Split the data

First 70% of the data will be the train and the rest of the 30% will be the test.

```python
X = df.values
```

```python
train, test = X[0:-round(len(X)*.3)], X[-round(len(X)*.3):]
```

Accumulate the history of the train data to compare predictions with.

```python
history = [x for x in train]
```

### Predictions

Predictions using walk-forward method with timestep of one.

```python
predictions = list()
for i in range(len(test)):
    predictions.append(history[-1])
    history.append(test[i])
```

Evaluate all the errors using root mean squared metric to summarize the strength of the model.

```python
rmse = np.sqrt(mean_squared_error(test, predictions))
print('RMSE: %.3f' % rmse)
# line plot of observed vs predicted
plt.plot(test, c='b')
plt.plot(predictions, c='r')
plt.title('Predictions vs True values')
plt.show()
```

Running the example prints the RMSE of about 38 weekly sale price for the forecasts on the test dataset. Although the rmse is a bit low, the one step walk-forward prediction is not practical and so we'll explore LSTM. 

## LSTM

### Transform data to stationary

The data has an increasing trend dependent on time and is not stationary. We will have to create a function to difference the data such that the observation from the previous time step $t_{-1}$ is subtracted from the current observation $t$ to get rid of the trend and we are left with the change in price.

```python
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return pd.Series(diff)
```

```python
raw_values = df.values
diff_values = difference(raw_values, 1)
```

### Transform data to be supervised learning

For LSTM models, we need to divide our data into input $X$ and output $y$ components by using the observation from the last time step $t_{-1}$ as the input and the observation at the current time step $t$ as the output. We will create a function that incorporates the shift() method to push the values down a sliding window of lag 4 and fill empty values with zeroes where the model will learn as the start of the series.

```python
def timeseries_to_supervised(data, lag=1):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    df.fillna(0, inplace=True)
    return df
```

```python
lag = 4
```

```python
supervised = timeseries_to_supervised(diff_values, lag)
supervised_values = supervised.values
```

### Split the data

First 70% of the data will be the train and the rest of the 30% will be the test.

```python
cutoff = round(.3*len(supervised_values))
```

```python
train, test = supervised_values[0:-cutoff], supervised_values[-cutoff:]
```

### Transform the scale of the data

Scale the data so that all the inputs will be on the same scale as the rest of the data.

```python
def scale(train, test):
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler.fit(train)
    train = train.reshape(train.shape[0], train.shape[1])
    train_scaled = scaler.transform(train)
    test = test.reshape(test.shape[0], test.shape[1])
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
```

```python
mmscaler, train_scaled, test_scaled = scale(train, test)
```

### Fit the Model

Reshaping input for model
- LSTM layer expects input to be in a 3D matrix
- Dimensions:
    - Samples: independent observations or rows of data.
    - Time steps: separate time steps of a given variable for an observation.
    - Features: separate measures observed at the time of observation.

Create a function for the model:
- recommended dropout of 0.1, one neuron, batch size of one (predicting one row), 1500 epochs
- batchsize and epochs determine how quickly the network learns the data and updates the weights
- stateful set to True to enable long term memory such that we can control when the state is cleared and the weights of the network is updated
- call the reset_state() to clear the state
- compile the network with specified loss function of 'mean-squared error' and gradient-based optimization 'adam'
- return the model

```python
def fit_lstm(train, batch_size, nb_epoch, neurons):
    X, y = train[:, :-1], train[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    model = Sequential()
    model.add(LSTM(neurons, dropout = 0.1, batch_input_shape=(batch_size, X.shape[1], X.shape[2]), stateful=True))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    for i in range(nb_epoch):
        model.fit(X, y, epochs=1, batch_size=batch_size, verbose=0, shuffle=False)
        model.reset_states()
    return model
```

Fit the model

```python
start = time.time()

lstm_model = fit_lstm(train_scaled, 1, 1500, 1)
train_reshaped = train_scaled[:, 0:lag].reshape(len(train_scaled), 1, lag)
lstm_model.predict(train_reshaped, batch_size=1)

end = time.time()
print(end-start)
```

### Testset Forecast

```python
def forecast_lstm(model, batch_size, X):
    X = X.reshape(1, 1, len(X))
    yhat = model.predict(X, batch_size=batch_size)
    return yhat[0,0]
```

We need to invert the scale process in order to take forecasts made on the scaled series back into their original scale.

```python
def invert_scale(mmscaler, X, value):
    new_row = [x for x in X] + [value]
    array = np.array(new_row)
    array = array.reshape(1, len(array))
    inverted = mmscaler.inverse_transform(array)
    return inverted[0, -1]
```

We need to invert the difference process in order to take forecasts made on the differenced series back into their original scale.

```python
def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]
```

Running the example below prints the expected and predicted values for each of the weeks months in the test dataset. The results of predicted are fairly close to the expected values.

```python
test_predictions = list()
for i in range(len(test_scaled)):
    X, y = test_scaled[i, 0:-1], test_scaled[i, -1]
    yhat = forecast_lstm(lstm_model, 1, X)
    yhat = invert_scale(mmscaler, X, yhat)
    yhat = inverse_difference(raw_values, yhat, len(test_scaled)+1-i)
    test_predictions.append(yhat)
    expected = raw_values[len(train) + i + 1]
    print('Week=%d, Predicted=%f, Expected=%f' % (i+1, yhat, expected))
```

### Evaluation

The example also prints the RMSE of all forecasts. The model shows an RMSE of 14.163 weekly sale price, which is slightly better than the persistence model that achieved an RMSE of 15.456 weekly saleprice.

```python
rmse = np.sqrt(mean_squared_error(raw_values[-cutoff:], test_predictions))
print('Test RMSE: %.3f' % rmse)

plt.figure(figsize=(15, 8))
plt.title(f'Jordan Bred 1 True Value vs Predictions', fontsize=25)
plt.plot(raw_values[-cutoff:], color='#5499C7', label='Expected')
plt.plot(test_predictions, color='#EFA74B', label='Predicted')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel(f'Date', fontsize=18)
plt.ylabel('Saleprice ($)', fontsize=18)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

### Forecast the next year

Create a function where the last predictions will be used as the input for next predictions.

```python
future_predictions = list()
X = test_scaled[-1, :-1]

for i in range(len(test_scaled)):
    yhat = forecast_lstm(lstm_model, 1, X)
    yhat = invert_scale(mmscaler, X, yhat)
    yhat = inverse_difference(raw_values, yhat, len(test_scaled)+1-i)
    future_predictions.append(yhat)
```

### Plot Forecasted Predictions

Add the new predictions with their respective weekly dates and create a dataframe and concat it to the dataframe of history of sales.

```python
curr_date = df.index[-1]
new_dates = []

for i in range(len(future_predictions)):
    curr_date = curr_date + timedelta(days=7)
    new_dates.append(curr_date)
```

```python
curr_df = pd.DataFrame(new_dates, columns=['date'])

old_df = pd.DataFrame(df)

curr_df.set_index(curr_df['date'], inplace=True)

curr_df['saleprice'] = future_predictions

final_df = pd.concat([old_df, curr_df.drop('date', axis=1)])
```

Below we plotted the combined historic and forecasted weekly saleprices.

```python
def basicPlot(df):
    plt.figure(figsize=(15, 8))
    plt.plot(final_df['saleprice'], color='#EFA74B')
    plt.title(f'Jordan Bred 1 Forecasted Average Weekly Sale Price', fontsize=25)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.xlabel(f'Date', fontsize=18)
    plt.ylabel('Saleprice ($)', fontsize=18)
    plt.grid()
    plt.show();
```

```python
basicPlot(final_df)
```

## Multiclass Classification Logistic Regression

We will be doing multiclass classification to sell the sneaker (0) because the price is 17.5% more than the mean, hold off on the sneaker(1) because the price is within 17.5% of the mean, or buy the sneaker(2) because the price is 25% less than the mean. The 17.5% is a big margin because StockX has 9.5% fees for all new-users(requires many consistent sales to upgrade account tier and access lower fees) and 3% processing fee totaling to 12.5% which leaves 5% profit.

### Create a label for the dataset

```python
avg_price = df.mean()
avg_price
```

```python
def action_of_customer(price):
    if price / 1.25 < avg_price:
        return 0
    if (price >= avg_price / 1.175) & (price <= avg_price * 1.175):
        return 1
    if price / 1.175 > avg_price:
        return 2
```

```python
train_labels = [action_of_customer(x) for x in df.values]
```

## Append labels to dataframe

```python
train_labels = pd.Series(train_labels)
```

```python
df = pd.concat([supervised, train_labels],axis=1)
df.drop(len(df)-1, inplace=True)
columns = ['lag_1','lag_2','lag_3','lag_4','lag_-1','action']
df.columns = columns
```

### Split the data

```python
X = df[columns[:-1]]
y = df[columns[-1]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.3,
                                                    random_state=42)
```

### Fit the model

```python
mul_lr = linear_model.LogisticRegression(multi_class='multinomial', solver='newton-cg').fit(X_train, y_train)
```

### Evaluate the model

```python
accuracy_score(y_train, mul_lr.predict(X_train))
```

```python
accuracy_score(y_test, mul_lr.predict(X_test))
```

```python
predictions = mul_lr.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```
